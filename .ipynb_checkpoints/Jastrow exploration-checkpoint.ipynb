# Methods

```python
def remove_html(entry):
    """
    Removes html and returns as dict.
    """
    entry2 = deepcopy(entry)
    
    obj_id = entry2.pop('_id')
    
    text = json.dumps(entry2)
    clean = re.sub(r'</*.+?>', ' ', text)
    
    as_dict = json.loads(clean)
    as_dict['_id'] = obj_id
    return as_dict
```

```python
def get_defs(entry):
    """
    Gets the definitions of a word as a lit of tuples. To be used on
    the entry only after HTML is removed.
    """
    senses = entry['content']['senses']
    defs = []
    for s in senses:
        if 'definition' not in s: continue
        e = re.sub(r'[a-zA-Z]{1,4}\.', '', s['definition'])
        e = e.replace(';', ',')
        e = re.sub(r'((\(|\[).*?(\)|\])|[^a-zA-Z,.\s\-])', '', e)
        print(e)
        d = e.split('.')[0]
        print(d)
        defs.append( tuple([w.strip() for w in d.split(',')]) )
    return defs
```
