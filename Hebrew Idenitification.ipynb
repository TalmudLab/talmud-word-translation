```python
for p in ber[:5]:
    for c in p:
        print(' '.join(c))
        no_punc = [char for char in c if char not in ',.?!\'\"״-—:;׳']
        tot = 0
        heb = 0
        for words in ngrams(no_punc, 4):
            tot += 1
            heb += 1 * (sum([w in heb_corpus for w in words]) > 2)
        code = 'u' if tot == 0 else ('m' if heb / tot > 0.85 else 'g')
        print(str(heb) + ' / ' + str(tot) + '\t' + code, end='\n\n')
```

# By Sentence Unit

Results: poor

```python
sentences = []
for p in ber:
    for c in p:
        sen = ''
        for t in c:
            if t in '״׳':
                continue
            elif t in '.!?:;—,':
                sentences += [sen.strip()]
                sen = ''
            else:
                sen += ' ' + t
sentences = [s for s in sentences if s != '']
```

```python
for s in sentences[:1000]:
    print(s)
    
    words = s.split()
    t_words = len(words)
    h_words = sum([(w in heb_corpus) for w in words])

    code = 'm' if h_words / t_words > 0.6 else 'g'
    print(str(h_words) + ' / ' + str(t_words) + '\t' + code, end='\n\n')
```

```python
brayta_starters = [('וּרְמִינְהוּ',), ('כִּדְתַנְיָא',), ('דְּתַנְיָא',), ('תַּנְיָא',), ('דִּתְנַן',), ('תָּנוּ', 'רַבָּנָן'), ('תָּנָא',)]
```

```python
five_pgs = [c for p in ber[5:10] for c in p]
```

```python
is_tanna = True
for c in five_pgs:
    print(' '.join(c))
    
    if not is_tanna:
        is_tanna = any([tuple(list(s) + [':']) in ngrams(c, len(s)+1) for s in brayta_starters])
        if is_tanna: print('START\tm', end='\n\n')
        else: print('g\n')
    
    else:
        no_punc = [char for char in c if char not in ',.?!\'\"״-—:;׳']
        tot_words = len(no_punc)
        heb_words = 0
        for w in no_punc:
            if w in heb_corpus:
                heb_words += 1
        code = 'm' if heb_words / tot_words > 0.6 else 'g'
        print(str(heb_words) + ' / ' + str(tot_words) + '\t' + code, end='\n\n')
        is_tanna = code == 'm'
```
