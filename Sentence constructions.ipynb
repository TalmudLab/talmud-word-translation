# Common words

```python
from collections import defaultdict
```

```python
occurences = defaultdict(lambda: 0)
for p in range(len(ber)):
    for c in ber[p]:
        for t in c:
            if t in ',.!?:;״()[]—׳…':
                continue
            occurences[t] += 1
occurences = [(occurences[k], k) for k in occurences]
```
